## Etapa 1 do CRISP-DM: Compreensão do Negócio <a name="1"></a>

**Este projeto trata de um problema relacionado à concessão de cartões de crédito**, apresentado no [Kaggle](https://www.kaggle.com/), uma plataforma que incentiva desafios de ciência de dados com prêmios em dinheiro para os vencedores. Você pode acessar o link original [aqui](https://www.kaggle.com/rikdifos/credit-card-approval-prediction).

O foco está em uma base de dados de solicitantes de cartão de crédito. **O principal objetivo é desenvolver um modelo preditivo capaz de identificar o risco de inadimplência**, definido como a ocorrência de um atraso de pagamento igual ou superior a 90 dias ao longo de um período de 12 meses, utilizando variáveis disponíveis no momento da avaliação de crédito (geralmente quando o cliente solicita o cartão).

Atividades do CRISP-DM:

- Objetivos do Negócio

>O propósito aqui é que o modelo sirva como uma ferramenta para o mutuário (o cliente), permitindo-lhe avaliar suas próprias decisões financeiras, ao contrário de servir diretamente à instituição de crédito.

- Objetivos da Modelagem

>O objetivo é claro: desenvolver o modelo preditivo mais eficaz possível para auxiliar o mutuário na tomada de decisões relacionadas ao crédito.

### Entendimento dos dados - Análise bivariada <a name="bivariada"></a>

```python
idade_quartis, bins = pd.qcut(x=df['idade'], q=4, retbins=True)

grupo_idade = df.groupby(idade_quartis, observed=False)

proporcao_inadimplentes = grupo_idade['mau'].mean()

fig, ax = plt.subplots()
proporcao_inadimplentes.plot(ax=ax, kind='line')

ax.set_xlabel('Idade em anos')
ax.set_ylabel('Proporção de inadimplentes')
ax.set_xticks(range(len(proporcao_inadimplentes.index)))
ax.set_xticklabels(proporcao_inadimplentes.index, rotation=45)

plt.tight_layout()
plt.show()
```

```python
tempo_emprego_quartis, bins = pd.qcut(x=df['tempo_emprego'], q=4, retbins=True)

grupo_tempo_emprego = df.groupby(tempo_emprego_quartis, observed=False)

proporcao_inadimplentes = grupo_tempo_emprego['mau'].mean()

fig, ax = plt.subplots(figsize=(10, 6))  
proporcao_inadimplentes.plot(kind='line', ax=ax, marker='o', linestyle='-')

ax.set_xlabel("Tempo de emprego em anos", fontsize=12)
ax.set_ylabel("Proporção de inadimplentes", fontsize=12)
ax.set_title("Proporção de Inadimplentes por Quartis de Tempo de Emprego", fontsize=14)
ax.set_xticks(range(len(proporcao_inadimplentes.index)))
ax.set_xticklabels(proporcao_inadimplentes.index, rotation=45, ha='right')  # Ajusta a orientação dos labels

ax.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()

plt.show()
```

## Etapa 4 e 5 Crisp-DM: Modelagem & Avaliação dos resultados <a name="4e5"></a>

<div style="text-align: right"
     
[Voltar ao índice](#Contents)

### Dividindo a base em treino e teste <a name="train_test_split"></a>

<div style="text-align: right"
     
[Voltar ao índice](#Contents)

```python
X = df.drop(columns='mau')
print('Quantidade de linhas e colunas das variáveis explicativas (X):', X.shape, '\n')

y = df['mau']
print('Quantidade de linhas da variável resposta (y):', len(y), '\n\n')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)

print('Quantidade de linhas e colunas de X_train (70%):', X_train.shape)
print('Quantidade de linhas e colunas de X_test (30%):', X_test.shape, '\n')

print('Quantidade de linhas de y_train (70%):', len(y_train))
print('Quantidade de linhas de y_test (30%):', len(y_test))
```

### RandomForestClassifier <a name="RFC"></a>

<div style="text-align: right"
     
[Voltar ao índice](#Contents)

```python
clf = RandomForestClassifier(n_estimators=5)
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
acc = metrics.accuracy_score(y_test, y_pred)*100
print('Acurácia: {0:.2f}%'.format(acc))
```

#### Matriz de confusão:

```python
tab = pd.crosstab(index = y_pred, 
                  columns = y_test)
tab
```

### DecisionTreeClassifier #1 <a name="DTC1"></a>

<div style="text-align: right"
     
[Voltar ao índice](#Contents)

```python
clf = DecisionTreeClassifier(random_state=100)
clf.fit(X=X_train, y=y_train)
clf
```

```python
plt.figure(figsize=(20, 10))
plot_tree(clf, filled=True, feature_names=X_train.columns, class_names=True, rounded=True)
plt.title("Árvore de Decisão")
plt.show()
```

#### Avaliação do modelo:

```python
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)*100
print(f'Acurácia do modelo: {accuracy:.2f}%')
```

#### Relatório de classificação

```python
print("\nRelatório de Classificação:")
print(classification_report(y_test, y_pred))
```

#### Matriz de confusão para a base de treino:

```python
ConfusionMatrixDisplay.from_estimator(estimator=clf, 
                                      X=X_train, 
                                      y=y_train, 
                                      display_labels=['aprovados', 'reprovados'])

plt.show()
```

#### Acurácia na base de treinamento:

```python
predict_train = clf.predict(X=X_train)
accuracy_score(y_true=y_train, 
               y_pred=predict_train)*100
```

```python
predict_test = clf.predict(X=X_test)

print('Valores preditos:') 
print(pd.Series(predict_test).value_counts(), '\n')

print('Valores reais:')
print(y_test.value_counts())
```

#### Matriz de confusão para a base de teste:

```python
ConfusionMatrixDisplay.from_estimator(estimator=clf, 
                                      X=X_test, 
                                      y=y_test, 
                                      display_labels=['aprovados', 'reprovados'])

plt.show()
```

#### Comparação da acurácia entre base de treinamento e base de teste:

```python
print('Acurácia da base de treinamento:', 
      accuracy_score(y_true=y_train, 
                     y_pred=predict_train), '\n')

print('Acurácia da base de teste:', 
      accuracy_score(y_true=y_test, 
                     y_pred=predict_test))
```

### DecisionTreeClassifier #2 <a name="DTC2"></a>

<div style="text-align: right"
     
[Voltar ao índice](#Contents)

```python
nova_clf = DecisionTreeClassifier(max_depth=10, 
                                  min_samples_leaf=5, 
                                  random_state=123)
nova_clf = nova_clf.fit(X=X_train, 
                        y=y_train)
nova_clf
```

```python
fig, axes = plt.subplots(nrows=2, 
                         ncols=2, 
                         figsize=(10,10))
fig.subplots_adjust(wspace=0.5)

ConfusionMatrixDisplay.from_estimator(estimator=nova_clf, 
                                      X=X_test, 
                                      y=y_test, 
                                      display_labels=['bons', 'maus'], 
                                      normalize=None, 
                                      ax=axes[0,0])
axes[0,0].set_title(label="normalize=None")

ConfusionMatrixDisplay.from_estimator(estimator=nova_clf, 
                                      X=X_test, 
                                      y=y_test, 
                                      display_labels=['bons', 'maus'], 
                                      normalize='true', 
                                      ax=axes[0,1])
axes[0,1].set_title(label="normalize='true'")

ConfusionMatrixDisplay.from_estimator(estimator=nova_clf, 
                                      X=X_test, 
                                      y=y_test, 
                                      display_labels=['bons', 'maus'], 
                                      normalize='pred', 
                                      ax=axes[1,0])
axes[1,0].set_title(label="normalize='pred'")

ConfusionMatrixDisplay.from_estimator(estimator=nova_clf, 
                                      X=X_test, 
                                      y=y_test, 
                                      display_labels=['bons', 'maus'], 
                                      normalize='all', 
                                      ax=axes[1,1])
axes[1,1].set_title(label="normalize='all'")

plt.show()
```

```python
print("Proporção de proponentes classificados como 'maus':",
      round(7/(1937+87+7)*100, 2), 
      '%')
```

#### Acurácia caso todos os contratos fossem classificados como 'bons':

```python
print("Valores reais:")
print(pd.Series(y_test).value_counts(), 
      '\n')

print("Todos os contratos da base teste classificados como 'bons':")
print(pd.Series(y_test.replace(to_replace=True, 
                               value=False)
               ).value_counts(), 
      '\n')

print('Acurácia:', 
      round(accuracy_score(y_true=y_test, 
                           y_pred=y_test.replace(to_replace=True, 
                                                 value=False)
                          ) * 100,
            2), 
      '%')
```
